```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axs = plt.subplots(1, 1, figsize=(7+3, 6))
cmap = 'jet_r'
levels = 100
df = pd.read_csv('results/q-rho-500.csv')
df = df.groupby(['Q','rho']).mean().reset_index()
matrix = df.pivot(index='rho', columns='Q', values='bestLength')
X=matrix.columns.values
Y=matrix.index.values
Z=matrix.values
axs.set_xlabel('Q')
axs.set_ylabel('rho')
cs = axs.contourf(X,Y,Z, levels, cmap=cmap)
plt.colorbar(cs, ax=axs)
plt.show()
```
